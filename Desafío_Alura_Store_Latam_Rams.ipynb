#Primeros pasos del desafío

###Importación de datos



```python
#Importamos la biblioteca de pandas para manipulación de datos

import pandas as pd

#URLs de los archivos CSV correspondientes a las 4 tiendas de Alura Store
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#Leyendo los archivos CSV y almacenamos cada tienda en DataFrames individuales
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

###Unificando los DataFrames en uno solo para facilitar comparaciones entre tiendas

```python
#Agregamos una columna que identifique a cada tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

#Unificamos todos los DataFrames en uno solo para facilitar el análisis comparativo, usando el método pd.concat
df_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
#Viendo el tamaño de df_tiendas
df_tiendas.shape
```

```python
#Visualizando los metadatos de los datos de todas las tiendas (df_tiendas)
df_tiendas.info()
```

###Convertimos la columna "Fecha de compra" a formato de fecha

```python
#Conversión usando pd.to_datetime; especificando el formato de fecha
df_tiendas['Fecha de Compra'] = pd.to_datetime(df_tiendas['Fecha de Compra'], format='%d/%m/%Y')
```

```python
#Observando si se realizó el cambio
df_tiendas.info()
```

#1. Análisis de facturación



```python
#Cálculo de ingreso por día en la tienda1
ingreso_por_tienda1 = tienda1.groupby('Fecha de Compra')['Precio'].sum().sort_values(ascending=False)

#Sumatoria de los ingresos para obtener el ingreso total de la tienda
ingreso_total_tienda1 = tienda1['Precio'].sum()

#Imprimir resultado
print(ingreso_total_tienda1)
```

```python
#Cálculo de ingreso total para todas las tiendas

ingresos_por_tienda = df_tiendas.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)

#Imprimir resultados de ingresos por tienda
print("Ingresos por tienda:")
print(ingresos_por_tienda)
```

```python
#Mostrar los ingresos con puntos como separador de miles, para una mejor visualización de las cifras

ingresos_por_tienda_formateados = ingresos_por_tienda.apply(lambda x: f'{x:,.2f}')
print("Ingresos por tienda:")
print(ingresos_por_tienda_formateados)
```

# 2. Ventas por categoría

###Ver elementos dentro de la categoría

```python
#Observamos cuántos elementos existen dentro de la columna de "categoría del producto", ordenado alfabéticamente

categorias_ordenadas = sorted(df_tiendas['Categoría del Producto'].unique())

#Imprimir resultado para observar
print(categorias_ordenadas)
```

```python
#Ver los elementos junto a la cantidad de veces que se vendieron en total (sin distinguir por tiendas)
conteo_categorias = df_tiendas['Categoría del Producto'].value_counts()

#Imprimir resultado para observar
print("Ventas por producto:")
print(conteo_categorias)
```

###Agruparlos para observar la cantidad de ventas por categoría del producto de cada tienda

```python
#Agrupamos por tienda y categoría del producto, y contamos
ventas_por_categoria = df_tiendas.groupby(['Tienda', 'Categoría del Producto'])['Producto'].count().reset_index()

#Renombramos la columna para mejor claridad
ventas_por_categoria.columns = ['Tienda', 'Categoría del Producto', 'Cantidad de Ventas']

#Reorganizamos los datos ordenando de mayor a menor, para visualizarlos mejor
ventas_por_categoria = ventas_por_categoria.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, False])

#Mostramos los resultados, separando las categorías por tienda
print("Ventas por producto:")
for tienda in ventas_por_categoria['Tienda'].unique():
    print(f"\n{'='*24} {tienda} {'='*24}") #Dando las separaciones
    print(ventas_por_categoria[ventas_por_categoria['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

###Sumatoria de cantidad total de ventas por tienda

```python
#Contamos cuántas ventas hay por tienda (una fila = una venta)
cantidad_ventas_por_tienda = df_tiendas['Tienda'].value_counts().sort_index()

#Convertimos a un DataFrame para poder añadir la fila total que deseamos
cantidad_ventas_por_tienda_df = cantidad_ventas_por_tienda.reset_index()
cantidad_ventas_por_tienda_df.columns = ['Tienda', 'Cantidad de Ventas']

#Agregamos una fila final con el total de todas las tiendas
total_general = pd.DataFrame([['Total general', cantidad_ventas_por_tienda.sum()]], columns=cantidad_ventas_por_tienda_df.columns)

#Concatenamos la tabla original con la fila total
ventas_con_total = pd.concat([cantidad_ventas_por_tienda_df, total_general], ignore_index=True)

#Mostramos el resultado
print("Cantidad de ventas totales por tienda:")
print(ventas_con_total)
```

# 3. Calificación promedio (media) de la tienda




```python
#Calculamos la calificación promedio de cada una de las tiendas; redondeado a 2 cifras decimales
calificacion_promedio_tienda = df_tiendas.groupby('Tienda')['Calificación'].mean().round(2)

#Mostramos el resultado
print("Calificación promedio por tienda:")
print(calificacion_promedio_tienda)
```

# 4. Productos más y menos vendidos

```python
#Agrupamos por tienda y producto, y contamos cuántas veces se vendió cada producto
ventas_por_producto = df_tiendas.groupby(['Tienda', 'Producto'])['Producto'].count().reset_index(name='Cantidad de Ventas')

#Ordenamos por tienda y luego por ventas descendente
ventas_por_producto = ventas_por_producto.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, False])
```

###Productos más vendidos de las tiendas

```python
#Obtenemos los 5 productos más vendidos por tienda
top5_productos = ventas_por_producto.groupby('Tienda').head(5).reset_index(drop=True)

#Imprimir resultados obtenido
print("Productos más vendidos por tienda:")
for tienda in top5_productos['Tienda'].unique():
    print(f"\n{'='*25} {tienda} {'='*25}") #Dando las separaciones
    print(top5_productos[top5_productos['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

###Productos menos vendidos de las tiendas

```python
#Ordenamos por tienda y cantidad de ventas, de forma ascendente
menos_vendidos = ventas_por_producto.sort_values(by=['Tienda', 'Cantidad de Ventas'], ascending=[True, True])

#Obtenemos los 5 productos menos vendidos por tienda
bottom5_productos = menos_vendidos.groupby('Tienda').head(5).reset_index(drop=True)

#Imprimir resultados
print("Productos menos vendidos por tienda:")
for tienda in bottom5_productos['Tienda'].unique():
    print(f"\n{'='*25} {tienda} {'='*25}") #Dando las separaciones
    print(bottom5_productos[bottom5_productos['Tienda'] == tienda].reset_index(drop=True)) #Mostrar resultados sin su índice
```

# 5. Envío promedio por tienda

```python
#Agrupamos por tienda y calculamos el promedio del valor de envío
envio_promedio = df_tiendas.groupby('Tienda')['Costo de envío'].mean().sort_values(ascending = False).round(2)

#Mostramos los resultados
print("Costo de envío promedio por tienda:")
print(envio_promedio)
```

#6.Gráficos para comprender mejor los datos analizados

```python
#Importando Matplotlib para graficar
import matplotlib.pyplot as plt

#Importan ticker de Matplotlib.ticker
import matplotlib.ticker as ticker
```

###Gráfico de barras: ingresos totales por tienda

```python
#Generando el gráfico
fig1, ax1 = plt.subplots(figsize=(10, 5))

#Dándole color al gráfico: fondo y fondo del área de lod ejes
fig1.patch.set_facecolor('#2b2b2b')
ax1.set_facecolor('#2b2b2b')

#Dándole sus especificaciones
ingresos_por_tienda.plot(kind='bar', color='#F5B041', ax=ax1)

#Ajustando títulos y sus labels
ax1.set_title("Ingresos totales por tienda", fontsize=16, color = 'white')
ax1.set_ylabel("Ingresos en millones de USD", fontsize=14, color = 'white')
ax1.set_xlabel("Tiendas", fontsize=14, color = 'white')

#Ajustando la inclinación de los labels del eje X
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=20, fontsize = 10, color = 'white')

#Ajustando tamaño de las cifras del eje Y
ax1.tick_params(axis='y', labelsize=10, colors = 'white')

#Ajustando el cuadriculado
ax1.grid(axis='y', linestyle='-', alpha=0.3, color = 'white')

#Colocando las cifras encima de cada barra, con separador de miles
ax1.bar_label(ax1.containers[0], fmt='{:,.2f}', color = 'white', fontsize = 10)

#Formatear los valores del eje Y con separador de miles
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

#Ajustar el rango del eje Y
ax1.set_ylim(0, 1_300_000_000)

#Mostrar gráfico
plt.tight_layout()
plt.show()
```
